```python
import cv2
 #載入級聯分類器
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
 
face_filename = 1
def detect(filename):
    img = cv2.imread(filename)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #RGB轉灰階
    faces = face_cascade.detectMultiScale(gray,               #臉部辨識 
                                          scaleFactor=1.2,   #scaleFactor尺度係數
                                          minNeighbors=3,)   #minNeighbors需要的鄰域數
                                         #minSize=(40,40),)  #minSize 設定數據搜尋的最小尺寸                   
    for (x,y,w,h) in faces:
        roi_gray = gray[y:y+h, x:x+w]                         
        eyes = eye_cascade.detectMultiScale(roi_gray,        #眼睛辨識 
                                            scaleFactor=1.3,
                                            minNeighbors=3,)
        
        img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2) #繪出臉部和眼睛方框
        for (ex,ey,ew,eh) in eyes:
            img = cv2.rectangle(img,(x+ex,y+ey),(x+ex+ew,y+ey+eh),(0,255,0),2)
        
        if len(eyes)>=2:                                   #至少辨識出兩顆眼睛才儲存
            f = cv2.resize(gray[y:y+h, x:x+w], (200, 200)) # 把影象全部格式化成200x200畫素
            global face_filename
            name = './P/%d.png' % face_filename  #將辨識結果圖片寫入P資料夾內並以png檔儲存
            cv2.imwrite(name,img)
            face_filename += 1
        else:
            pass
    print('Working with %s' % filename)
```

```python
import os
 
img_list = []  #把跟程式碼同級資料夾層的各種格式圖片儲存成list
for fileNames in os.walk("."):
    filename_extension = ['.jpg', '.png', '.jpeg', '.JPG', '.gif']
    for fe in filename_extension:
        for fileName in fileNames[-1]:
            if fileName.endswith(fe):
                img_list.append(fileName)
 
for img in img_list:
    detect(img) 
```
